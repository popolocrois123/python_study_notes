### これでは文字列を分割できないのでlistを用いる

```python
# 配列を用意
n = int(input())
arr = [0] * n

# 整数を受け取って
line = list(input())
print(line)
for i in range(n):
    query = int(line[i])
    arr[i] = query

# それを先頭から順にしまう
for j in range(n):
    print(arr[j])
```

### スタックとキューの要素の追加push と取り出しpop 


```python
# 配列を用意
n = int(input())
# 配列の末尾
last = 0
arr = [0] * 256

for i in range(n):
    query = input().split()
    # もしpushだった場合整数を受け取って配列の末尾にしまう
    if query[0] == "push":
        x = int(query[1])
        arr[last] = x
        last += 1

for j in range(last):
    print(arr[j])
```

```python
# クエリの数
Q = int(input())
last = 0
# クエリのリストをあらかじめ作っておく
query = [0] * 256
# 最後にprintする回数のカウント
count = 0

# 各クエリの内容を取得して配列に追加か変更
for i in range(Q):
    input_query = input().split()
    if input_query[0] == "1":
        query[last] = input_query[1]
        last += 1
        count += 1
        # print(f"テスト{query[0:3]}")
    elif input_query[0] == "2":
        old_index = int(input_query[1]) - 1
        new_index = int(input_query[2]) - 1
        query[old_index], query[new_index] = query[new_index], query[old_index]

for j in range(count):
    print(query[j])
        
```
