

お客さんが一列に並んで先頭の人がお店に入る


長さ５のエスカレーターを１０秒間しゅみゅレーション
各秒について

```python
# 長さ５のエスカレーター
escalator_length = 5

queue = []
person_on_escalator = 0

# １０秒間回す
# 1, 3, 7秒の時にお客さんが入ってくる
for time in range(1, 10):
    if time == 1:
        queue.append(2)
        person_on_escalator += 2
    elif time == 3:
        queue.append(5)
        person_on_escalator += 5
    elif time == 7:
        queue.append(2)
        person_on_escalator += 2
    else:
        queue.append(0)

    if len(queue) > escalator_length:
        person_on_escalator -= queue.pop(0)
        
    print(person_on_escalator)
    # print(queue)

```

この問題は少し難しいですが、キューを用いて実装してみましょう。なお、サンプル 1 のケースでは以下のように社員がエスカレーターに乗ります。





【入力される値】

N K
A_1 ... A_N



【条件】

すべてのテストケースにおいて、以下の条件をみたします。

・ N, A_i は 1 以上 50,000 未満
・ 1 ≦ A_1 < ... < A_N < 50,000
・ K は 1 以上 100 未満

社員 i がエスカレーターに乗った直後に、エスカレーター上にいる人数 E_i を改行区切りで出力してください。末尾には改行を入れ、余計な文字、空行を含んではいけません。

E_1
...
E_N
